```python

import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
import os
from sklearn.model_selection import train_test_split
import json
from math import sin, cos
import imutils
from skimage.transform import rotate

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset
from torchvision import datasets, models, transforms
from torchvision.transforms.autoaugment import AutoAugmentPolicy


import cv2
from PIL import Image

import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
cloud_paths = sorted(glob("E:\\uhin\\Proryv\\moskva\\MIPT\\data\\train\\cloudy\\cloudy_image\\*"))

orig = cv2.imread("data/original_2.tiff")

256*2**0.5

test_f = open('data/train/cloudy/test_list.txt', 'w')
train_f = open('data/train/cloudy/train_list.txt', 'w')
i = 0

for path in cloud_paths:
    ind = int(path.split('\\')[-1].split(".")[-2])
    path_json = "\\".join(['data\\train\\json']+[str(ind)+'.json'])
    with open(path_json, 'r') as f:
        params = json.load(f)
    print(params)
    cx = (params['left_top'][0] + params['right_bottom'][0])//4
    cy = (params['left_top'][1] + params['right_bottom'][1])//4
    angle = int(params['angle'])
    patch_big = orig[cy-363:cy+363,cx-363:cx+363]
#         print(c_y_c_x)
#         print(patch_big.shape)
    new_patch = (rotate(patch_big, angle)[363-256:2*363-(363-256), 363-256:2*363-(363-256), :]*255).astype('int')
    cv2.imwrite(f'data/train/cloudy/ground_truth/{ind:04d}.png', new_patch)
    train_f.write(f'{ind:04d}.png\n')
#     if i%rel == 0:
#         test_f.write(f'{ind:04d}.png\n')
#     else:
#         train_f.write(f'{ind:04d}.png\n')
    i+=1

train_f.close()

test_f.close()
```
